# Overview and Motivation:

  As international students, obtaining and maintaining our legal status here in the US are always our primary concern, especially after we graduate from college or graduate school. H-1B is the visa that any international students would need to obtain once we have found a job, but finding a job in no way guarantees a valid H-1B status. After we file an application for a H-1B visa, the applications are evaluated by USCIS, and only some subsets are approved each year. We normally think of the process as throwing our application into a lottery-like pool and each year, a subset of the applications will be drawn. However, the drawing process is not entirely random and unbiased. For example, people with PhD degrees have a higher chance than those with a masters, and master degree candidates have better off than people with bachelor’s degree. These are the limited parameters we know that affect our chance of being approved after filing an application. There are always rumors going around saying people with certain types of job or employers that are more well-known have a better chance of getting approved. Therefore,we wish to learn more about what affect does various parameters such as wage, types of jobs, time of application, employer, etc has on getting approval so that we will be able to predict whether a person will be approved or not based given a specific case.

Some background information about H-1B visa: H-1B is a non-immigrant VISA in the U.S. that allows U.S. employers to temporarily employ foreign workers in specialty occupations. According to U.S. Code for Admission of nonimmigrants, specialty occupations is defined as “theoretical and practical application of a body of highly specialized knowledge, and attainment of a bachelor’s or higher degree in the specific specialty (or its equivalent) as a minimum for entry into the occupation in the United States.” [Legal Information about H-1B can be found here](http://www.law.cornell.edu/uscode/html/uscode08/usc_sec_08_00001184----000-.html)

```python
def combine_status(df):
    """
    Helper function for data wrangling after reading in excel files
    Returns a list containing the combined value
    from the two columns 'APPROVAL_STATUS' and 'WITHDRAWN'
    
    If APPROVAL_STATUS is CERTIFIED and WITHDRAWN is Y, then the new status is CERTIFIED-WITHDRAWN
    If APPROVAL_STATUS is CERTIFIED and WITHDRAWN is N, then the new status remains as CERTIFIED
    If APPROVAL_STATUS is DENIED and WITHDRAWN is Y, then the new status is WITHDRAWN
    If APPROVAL_STATUS is DENIED and WITHDRAWN is N, then the new status remains as DENIED
    """    
    status_list = []
    sub = df[['WITHDRAWN','APPROVAL_STATUS']]
    for i in range(sub.shape[0]):
        if((sub['APPROVAL_STATUS'][i].encode('UTF-8') == 'CERTIFIED') & (sub['WITHDRAWN'][i].encode('UTF-8')=='Y')):# STATUS is CERTIFIED-WITHDRAWN
            status_list.append(u'CERTIFIED-WITHDRAWN')
        elif((sub['APPROVAL_STATUS'][i].encode('UTF-8') == 'DENIED') & (sub['WITHDRAWN'][i].encode('UTF-8')=='Y')): # STATUS is WITHDRAWN
            status_list.append(u'WITHDRAWN')
        else:
            status_list.append(sub['APPROVAL_STATUS'][i])
    return status_list
```

```python
def concat_addr(df,col_list):
    """
    Helper function for data wrangling after reading in excel files
    Returns a list containing the concatenated value
    from the two columns 'ADDRESS1' and 'ADDRESS2'
    """    
    add_list = []
    sub = df[[col_list[0],col_list[1]]]
    for i in range(sub.shape[0]):
        if(type(sub[col_list[1]][i]) == float):
            add_list.append(sub[col_list[0]][i])
        else:
            add_list.append(str(sub[col_list[0]][i])+' '+str(sub[col_list[1]][i]))
    return add_list
```

```python
def data_wrangle(filename):
    """
    Returns a pandas DataFrame object containing
    the wrangled data returned from the given filename
    
    In particular, this function uniforms the dataframes
    to have the union/intersection of all possible columns
    from all datasets
    """   
    uniform_col_name = [u'LCA_CASE_NUMBER', u'STATUS', u'LCA_CASE_SUBMIT', u'DECISION_DATE', u'VISA_CLASS', u'LCA_CASE_EMPLOYMENT_START_DATE', u'LCA_CASE_EMPLOYMENT_END_DATE', u'LCA_CASE_EMPLOYER_NAME', u'LCA_CASE_EMPLOYER_ADDRESS', u'LCA_CASE_EMPLOYER_CITY', u'LCA_CASE_EMPLOYER_STATE', u'LCA_CASE_EMPLOYER_POSTAL_CODE', u'LCA_CASE_SOC_CODE', u'LCA_CASE_JOB_TITLE', u'LCA_CASE_WAGE_RATE_FROM', u'LCA_CASE_WAGE_RATE_TO', u'LCA_CASE_WAGE_RATE_UNIT', u'FULL_TIME_POS', u'TOTAL_WORKERS', u'LCA_CASE_WORKLOC1_CITY', u'LCA_CASE_WORKLOC1_STATE', u'PW_1', u'PW_SOURCE_1', u'OTHER_WAGE_SOURCE_1', u'YR_SOURCE_PUB_1', u'LCA_CASE_WORKLOC2_CITY', u'LCA_CASE_WORKLOC2_STATE', u'PW_2', u'PW_SOURCE_2', u'OTHER_WAGE_SOURCE_2', u'YR_SOURCE_PUB_2']
    df = pd.io.excel.read_excel(filename, index_col=None)
    if(len(df.columns) == 40):## meaning this is the 2008 dataset, rearranging and trimming of columns needed 
        # 1. Drop two columns that are completely empty
        if((df['CERTIFIED_BEGIN_DATE'].isnull().all()) & (df['CERTIFIED_END_DATE'].isnull().all())):
            df.drop(['CERTIFIED_BEGIN_DATE','CERTIFIED_END_DATE'], axis=1, inplace=True)
        # 2. Concatenate columns 'ADDRESS1' and 'ADDRESS2'
        df['ADDRESS'] = concat_addr(df,['ADDRESS1','ADDRESS2'])
        # 3. Combine columns 'APPROVAL_STATUS' and 'WITHDRAWN'
        df['STATUS'] = combine_status(df)
        # 4. Finally rearrange the columns into the correct order, then rename the columns to match uniform_col_name
        df = df[['CASE_NO','STATUS','SUBMITTED_DATE','DOL_DECISION_DATE','PROGRAM','BEGIN_DATE','END_DATE','NAME','ADDRESS','CITY','STATE','POSTAL_CODE','JOB_CODE','JOB_TITLE','WAGE_RATE_1','MAX_RATE_1','RATE_PER_1','PART_TIME_1','NBR_IMMIGRANTS','CITY_1','STATE_1','PREVAILING_WAGE_1','WAGE_SOURCE_1','OTHER_WAGE_SOURCE_1','YR_SOURCE_PUB_1','CITY_2','STATE_2','PREVAILING_WAGE_2','WAGE_SOURCE_2','OTHER_WAGE_SOURCE_2','YR_SOURCE_PUB_2']]
        df.columns = uniform_col_name
    elif(len(df.columns) == 39):## meaning this is the 2009 part 1 dataset, rearranging and trimming of columns needed
        # 1. Concatenate columns 'ADDRESS1' and 'ADDRESS2'
        df['ADDRESS'] = concat_addr(df,['EMPLOYER_ADDRESS1','EMPLOYER_ADDRESS2'])
        # 2. Combine columns 'APPROVAL_STATUS' and 'WITHDRAWN'
        df['STATUS'] = combine_status(df)
        # 3. Finally rearrange the columns into the correct order
        df = df[['CASE_NO','STATUS','SUBMITTED_DATE','DOL_DECISION_DATE','PROGRAM_DESIGNATION','BEGIN_DATE','END_DATE','EMPLOYER_NAME','ADDRESS','EMPLOYER_CITY','EMPLOYER_STATE','EMPLOYER_POSTAL_CODE','OCCUPATIONAL_CODE','JOB_TITLE','WAGE_RATE_1','MAX_RATE_1','RATE_PER_1','PART_TIME_1','NBR_IMMIGRANTS','CITY_1','STATE_1','PREVAILING_WAGE_1','WAGE_SOURCE_1','OTHER_WAGE_SOURCE_1','YR_SOURCE_PUB_1','CITY_2','STATE_2','PREVAILING_WAGE_2','WAGE_SOURCE_2','OTHER_WAGE_SOURCE_2','YR_SOURCE_PUB_2']]
        df.columns = uniform_col_name
    else: # for all other dataframs
        # Drop columns 'LCA_CASE_SOC_NAME',PW_UNIT_1,PW_UNIT_2, and LCA_CASE_NAICS_CODE
        # because it does not exist in dataset 2008 and 2009
        df.drop(['LCA_CASE_SOC_NAME','PW_UNIT_1','PW_UNIT_2','LCA_CASE_NAICS_CODE'], axis=1, inplace=True)
    return df
```

```python
def find_indicies(ls,key):
    """
    Returns a list of indicies to ls that contains key
    """   
    result = []
    for ix in range(len(ls)):
        if(ls[ix].find(key) > 0): # find the index that contains the 
            result.append(ix)
    return result
```

### Exploring Trends in Final Decision (Before more sophisticated data cleaning):

First, we will expore some trends throughout the years to gain an insight into the data. Start with trend in status.

```python
# Helper Function
def trend_info(df):
    """
    Returns a list of information that will show trends throughout the years computed from a given dataset
    0. total number of applicants = total number of cases
    1. certified rate = # of(certified+certified-withdrawn)/total number of cases
    2. withdrawn rate = # of withdrawn/total number of cases
    3. certified-withdrawn rate = # of certified-withdrawn/(# of certifed+certfied-withdrawn)
    4,5. certified-denied ratio = # of(certified+certified-withdrawn):# of denied (in decimal and in fraction)
    """   
    Status = np.array(df['STATUS'])
    N = len(Status)
    # create a list to store status information using numerical values
    ls = []
    for i in range(N):
        if(Status[i].encode('UTF-8') == 'DENIED'):
            ls.append(0)
        elif(Status[i].encode('UTF-8') == 'CERTIFIED'):
            ls.append(1)
        elif(Status[i].encode('UTF-8') == 'CERTIFIED-WITHDRAWN'):
            ls.append(2)
        else: ##WITHDRAWN
            ls.append(3)
    result = []
    # Element 0, total number of applicant
    result.append(N)
    # convert numbers to float to get decimal results
    N = float(N)
    c = float(ls.count(1)+ls.count(2))
    # 1st element, append the rounded up accpetance rate
    result.append(round(c/N,4))
    # 2nd element, append the rounded up withdrawn rate
    result.append(round(ls.count(3)/N,6))
    # 3rd element, append the rounded up certified-withdrawn rate
    result.append(round(ls.count(2)/N,4))
    # 4th element, append the decimal value of acceptance-denial ratio
    acc_den = c/ls.count(0)
    result.append(acc_den)
    # 5th element, append an interger ratio (a Fraction) of acceptance-denial ratio
    r = Fraction(acc_den).limit_denominator(1)
    result.append(r)
    return result
```

#### Changing to 0 and 1

```python
def to_binary(df, column_name):
    """
    Return the df with an added column of 0's and 1's converted from column_name. 
    Convert the column 'column_name' of the df into binary.
    Binary value is set to 0 if the value in the column is null and 1 otherwise.
    """
    new_col = 'num_'+column_name
    df[new_col] = 0
    null_idx = df[df[column_name].isnull()].index
    notnull_idx = df[df[column_name].notnull()].index           
    df[new_col][null_idx] = 0
    df[new_col][notnull_idx] = 1
```
